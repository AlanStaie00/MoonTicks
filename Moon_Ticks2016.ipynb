```python
# What we need
import pandas as pd
import numpy
%matplotlib inline
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Open and Read file
moon_2016 = "moon_2016.csv"
moon_2016_df = pd.read_csv(moon_2016)
moon_2016_df.head()
```

```python
#Convert date format
moon_2016_df['date'] = pd.to_datetime(moon_2016_df['date'])
```

```python
#Pull out New Moon dates
moon_2016_df.loc[moon_2016_df.phaseid ==1]
```

```python
#Pull out 1st Qtr dates
moon_2016_df.loc[moon_2016_df.phaseid ==2]
```

```python
#Pull out Full Moon dates
full_moon = moon_2016_df.loc[moon_2016_df.phaseid ==3]
full_moon
```

```python
#Pull out Last Qtr dates
moon_2016_df.loc[moon_2016_df.phaseid ==4]
```

```python
#Plot
moon_2016_df.plot (x='date', color='white', visible=False, legend=None)
plt.xticks(full_moon.date)
plt.xticklabels=(full_moon.date)

moon_2016_df['date'] = pd.to_datetime(moon_2016_df['date'])

plt.tick_params(axis='x', direction='in', length=195, color='red', width="2")
plt.show
```
